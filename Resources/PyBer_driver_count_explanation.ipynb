### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
lake_jonathanshire_df = pyber_data_df.loc[pyber_data_df.city == "Lake Jonathanshire"]
lake_jonathanshire_df
```

#### Each ride for a given city results in each row having the total driver count for that city continuously printed. So the actual driver count for just that city is the mean of the driver count column 

```python
lake_jonathanshire_driver_count = lake_jonathanshire_df.driver_count.mean()
lake_jonathanshire_driver_count
```

#### So the reason that this code produces the total number of drivers is because when you `groupby(["city"])` and take the `mean()["driver_count"]` you end up with each cities total driver count like the result produced for Lake Jonathanshire above. And when you sum those up, you end up with the total number of drivers in a given city type.

```python
# Create the Urban city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
# Get the average number of drivers for each urban city.
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count_df = pd.DataFrame(urban_driver_count)
urban_driver_count_df
print(f'The number of urban drivers is: {urban_driver_count.sum()}')
```

# OR, you just get the driver count before the merge from the `city_data_df`

```python
# 2. Get the total drivers for each city type
total_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
print(total_driver_count)
```
